```
import pandas as pd
import random
```

```
# Carregando o dataset e escolhendo uma linha aleatória
df = pd.read_csv("telemetry_anomalies_dataset.csv")
line = random.randint(0, len(df) - 1)
```

```
def rocket_temp(x, y):
    if (18 <= x <= 35) and (-5 <= y <= 30):
        rocket_temp = 200
        print('Temperaturas estão Ok')
    else:
        print('Temperaturas inadequadas')
        rocket_temp = 400
    return rocket_temp
```

```
def rocket_integrity(x):
    if x == 1:
        rocket_integrity = 200
        print('A integridade estrutural está OK')
    else:
        rocket_integrity = 400
        print('A integridade estrutural está inadequada')
    return rocket_integrity
```

```
def rocket_tank(x):
    if (95 <= x <= 145):
        rocket_tank = 200
        print('A pressão do tanque está Ok')
    else:
        rocket_tank = 400
        print('A pressão do tanque está inadequada')
    return rocket_tank
```

```
def rocket_module(x, y):
    if x == 1 and y == 1:
        rocket_module = 200
        print('O status dos módulos críticos estão OK')
    else:
        rocket_module = 400
        print('Os módulos críticos estão instáveis')
    return rocket_module
```

```
# Analise Energetica
def rocket_energy_kwh(x, y, z):
    rocket_energy_kwh = (x * y * (z / 100)) / 1000
    rocket_energy_kwh = round(rocket_energy_kwh, 3)
    return rocket_energy_kwh
```

```
def rocket_power(power, min):
    rocket_power = power / (min / 60)
    rocket_power = round(rocket_power, 3)
    return rocket_power
```

```
def is_safe(internal_temp, external_temp, battery_v, battery_soc, pressure, struct, crit, telem):
    if not (20 <= internal_temp <= 30): return False
    if external_temp < 0: return False
    if battery_v < 48: return False
    if battery_soc < 80: return False
    if not (95 <= pressure <= 135): return False 
    if struct != 1: return False
    if crit != 1: return False
    if telem != 1: return False
    return True
```

```
def launch():
    # Aqui está a correção: chamamos is_safe passando os dados da linha sorteada
    seguro_para_lancamento = is_safe(
        internal_temp=df.loc[line, "internal_temp_c"],
        external_temp=df.loc[line, "external_temp_c"],
        battery_v=df.loc[line, "battery_voltage_v"],
        battery_soc=df.loc[line, "battery_soc_percent"],
        pressure=df.loc[line, "tank_pressure_bar"],
        struct=df.loc[line, "structural_integrity"],
        crit=df.loc[line, "critical_modules_status"],
        telem=df.loc[line, "telemetry_link_status"]
    )
    
    print("-" * 30)
    if seguro_para_lancamento:
        print(f' O foguete {line} está PRONTO para lançamento (GO)')
    else:
        print(f' O foguete {line} NÃO poderá ser lançado (NO-GO)')
    print("-" * 30)
```

```
# ==========================================
# Execução do Script e Prints
# ==========================================
print(f'\nA linha {line} está sendo analisada agora, este é o dia {df.loc[line, "timestamp"]} \n')
```

```
rocket_temp(df.loc[line, "internal_temp_c"], df.loc[line, "external_temp_c"])
rocket_integrity(df.loc[line, "structural_integrity"])
rocket_tank(df.loc[line, "tank_pressure_bar"])
rocket_module(df.loc[line, "critical_modules_status"], df.loc[line, "telemetry_link_status"])
```

```
print("\n--- Análise Energética ---")
print(f'Capacidade energetica total: {df.loc[line,"power_load_kw"]}Kw')
print(f'Porcentagem de Kwh usados: {rocket_energy_kwh(df.loc[line,"battery_voltage_v"], df.loc[line,"battery_capacity_ah"], df.loc[line,"battery_soc_percent"])}%')
print(f'Consumo de energia estimado na decolagem: {rocket_power(df.loc[line, "energy_available_kwh"], df.loc[line, "estimated_autonomy_min"])}Kwh')
print(f'Perda energetica: {df.loc[line,"energy_loss_percent"]}%')
```

```
# Chamando a função de lançamento
launch()
```
